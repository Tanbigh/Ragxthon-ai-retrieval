```python
from retriever import create_vector_store

# vector database তৈরি
vectorstore = create_vector_store()

# demo queries
queries = [
    "What is Artificial Intelligence?",
    "What is Machine Learning?",
    "What is DBMS?"
]

for q in queries:
    docs = vectorstore.similarity_search(q)

    print("=================================")
    print("Question:", q)
    print("\nAnswer:")
    print(docs[0].page_content)
    print("=================================\n")
```

```python
from retriever import create_vector_store

vectorstore = create_vector_store()

queries = [
    "What is Artificial Intelligence?",
    "What is Machine Learning?",
    "What is DBMS?"
]

for q in queries:
    docs = vectorstore.similarity_search(q)


    answer = docs[0].page_content.split(".")[0] + "."

    print("Question:", q)
    print("Answer:", answer)
    print()
```
